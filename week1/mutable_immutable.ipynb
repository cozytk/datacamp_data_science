변수를 대입하는 과정이 '복사'를 하는 것이 아니라 '참조'를 하는 과정임.

그저 해당하는 메모리에 이름이 하나 더 생긴 것.

```
import sys
a = ['a','b','c']
print(sys.getrefcount(a))
b = a
print(sys.getrefcount(a))
b = a.copy()
print(sys.getrefcount(a))
```

**메모리의 주소값을 출력해보자!**

`id()`

```
help(id)
```

```
x = ["a", "b", "c"]
y = x
y[1] = "z"
```

```
print(id(x))
print(id(y))
```

```
x = ["a", "b", "c"]
print(id(x))
print(id(list(x)))
# x[:], x.copy()
y = list(x)
y[1] = 'z'
```

```
print(id(x))
print(id(y))
```
